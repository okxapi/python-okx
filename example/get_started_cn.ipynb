# 开始现货交易

### 现货交易 - 简单/单币种保证金模式

```python
import okx.Trade as Trade

flag = "1"  # 实盘：0, 模拟盘：1

tradeAPI = Trade.TradeAPI(api_key, secret_key, passphrase, False, flag)
```

#### 下限价单

```python
# 限价单
result = tradeAPI.place_order(
    instId="BTC-USDT",
    tdMode="cash",
    side="buy",
    ordType="limit",
    px="19000",
    sz="0.01"
)
print(result)

if result["code"] == "0":
    print("下单成功，订单 ID = ",result["data"][0]["ordId"])
else:
    print("下单失败，错误码 = ",result["data"][0]["sCode"], ", 错误信息 = ", result["data"][0]["sMsg"])
```

#### 下市价单

```python
# 市价单
result = tradeAPI.place_order(
    instId="BTC-USDT",
    tdMode="cash",
    side="buy",
    ordType="market",
    sz="100",
)
print(result)
```

#### 使用 tgtCcy=quote_ccy 下单（仅适用于现货）

```python
# 市价单
result = tradeAPI.place_order(
    instId="BTC-USDT",
    tdMode="cash",
    side="buy",
    ordType="market",
    sz="100",
    tgtCcy="quote_ccy" # 这决定 sz 参数的单位。默认值为 base_ccy
)
print(result)
```

#### 使用您自己的 clOrdId 下单

```python
# 市价单
result = tradeAPI.place_order(
    instId="BTC-USDT",
    tdMode="cash",
    side="buy",
    ordType="market",
    sz="100",
    clOrdId="003" # 您可以定义自己的客户端订单 ID
)
print(result)
```

### 现货交易 - 多币种/投资组合保证金模式

```python
# 全仓现货交易
result = tradeAPI.place_order(
    instId="BTC-USDT",
    tdMode="cross",
    side="buy",
    ordType="limit",
    px="1000",
    sz="0.01"
)
print(result)

if result["code"] == "0":
    print("下单成功，订单 ID = ",result["data"][0]["ordId"])
else:
    print("下单失败，错误码 = ",result["data"][0]["sCode"], ", 错误信息 = ", result["data"][0]["sMsg"])
```

### 如果您希望在价格达到某个水平时下单，您可以下算法订单

```python
result = tradeAPI.place_algo_order(
    instId="BTC-USDT",
    tdMode="cash",
    side="buy",             # 买入
    ordType="trigger",      # 订单类型
    sz="100",               # 订单数量：100USDT
    triggerPx="10000",      # 触发价格
    orderPx="-1",           # 委托价格。当 orderPx=-1 时，订单将作为市价单下单
    triggerPxType="last"    # 触发价格类型。last：最新成交价
)
print(result)
```

## 您也可以使用止盈止损订单来卖出账户中的币种

```python
result = tradeAPI.place_algo_order(
    instId="BTC-USDT",
    tdMode="cash",
    side="sell",                # 卖出
    ordType="conditional",      # 单向止盈止损
    sz="0.01",                  # 订单数量：0.01BTC
    tpTriggerPx="30000",        # 止盈触发价格
    tpOrdPx="-1",               # 止盈委托价格。设置为 -1 时，订单将作为市价单下单
    tpTriggerPxType="last"      # 止盈触发价格类型。last：最新成交价
)
print(result)
```
